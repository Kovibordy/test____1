##Задание для гениев *
Агрегируйте числовой признак по категориальному признаку с использованием метода **без использования** библиотеки pandas, и вычислите медиану и среднее значение для каждой категории.

```python
#Агрегация

numeric_values = [10, 15, 12, 20, 8, 25, 18, 14, 22]
categorical_values = ['A', 'B', 'A', 'B', 'A', 'C', 'B', 'C', 'C']

# Создаем словари для хранения значений и счетчиков для каждой категории
category_sum = {}
category_count = {}

# Проходим по спискам значений и категорий и агрегируем данные
for i in range(len(numeric_values)):
    category = categorical_values[i]
    value = numeric_values[i]

    if category in category_sum:
        category_sum[category] += value
        category_count[category] += 1
    else:
        category_sum[category] = value
        category_count[category] = 1

# Вычисляем медиану и среднее значение для каждой категории
category_median = {}
category_mean = {}

for category in category_sum:
    category_median[category] = category_sum[category] / category_count[category]
    category_mean[category] = category_sum[category] // category_count[category]  # целочисленное деление для среднего значения

# Вывод результатов
for category in category_median:
    print(f"Категория {category}:")
    print(f"Медиана: {category_median[category]}")
    print(f"Среднее значение: {category_mean[category]}")
    print()
```
